## Plot datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Granularity: 250ms

```python
import matplotlib.ticker as plticker
```

```python
data_250ms = pd.read_csv('../data/aggregated/aggregated_250ms.csv', index_col='attr_time')
```

```python
data_250ms.head()
```

```python
data_250ms.describe()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20,8))
axs[0].plot(data_250ms[['attr_x', 'attr_y', 'attr_z']])
axs[1].plot(data_250ms[['attr_azimuth', 'attr_pitch', 'attr_roll']])

loc = plticker.MultipleLocator(base=35000.0) # this locator puts ticks at regular intervals
axs[0].xaxis.set_major_locator(loc)
# axs[0].tick_params(labelrotation=45)
axs[1].xaxis.set_major_locator(loc)
# axs[1].tick_params(labelrotation=45)
plt.savefig('../figures/data_250ms.pdf') 
plt.show()
```
